```python
colombia_page.add_text(
    f"""
    <div style="
        background:#EEF7F0;
        color:#1B4332;
        border:1px solid #1B433222;
        padding:14px;
        border-radius:12px;
        width:100%; height:100%;
        box-sizing:border-box;
        line-height:1.5;
        overflow:hidden;  /* Cambiado de auto a hidden */
        font-family:Arial;
        display:flex;  /* A√±adido */
        flex-direction:column;  /* A√±adido */
    ">
        
    <div class="tabs-group" id="groupB" style="height:100%; display:flex; flex-direction:column;">
        <div class="tab-buttons">
            <button onclick="openTabGroup('groupB', 'tab1_')">GDP</button>
            <button onclick="openTabGroup('groupB', 'tab2_')">Social support</button>
            <button onclick="openTabGroup('groupB', 'tab3_')">Life expectancy</button>
            <button onclick="openTabGroup('groupB', 'tab4_')">Freedom</button>
            <button onclick="openTabGroup('groupB', 'tab5_')">Generosity</button>
            <button onclick="openTabGroup('groupB', 'tab6_')">Corruption</button>
        </div>

        <div id="tab1_" class="tab-content-secondary active" style="flex:1; min-height:0;">
            <div id="plot1_" class="plotly-container" style="width:100%; height:100%;"></div>
        </div>

        <div id="tab2_" class="tab-content-secondary" style="flex:1; min-height:0;">
            <div id="plot2_" class="plotly-container" style="width:100%; height:100%;"></div>
        </div>

        <div id="tab3_" class="tab-content-secondary" style="flex:1; min-height:0;">
            <div id="plot3_" class="plotly-container" style="width:100%; height:100%;"></div>
        </div>

        <div id="tab4_" class="tab-content-secondary" style="flex:1; min-height:0;">
            <div id="plot4_" class="plotly-container" style="width:100%; height:100%;"></div>
        </div>

        <div id="tab5_" class="tab-content-secondary" style="flex:1; min-height:0;">
            <div id="plot5_" class="plotly-container" style="width:100%; height:100%;"></div>
        </div>

        <div id="tab6_" class="tab-content-secondary" style="flex:1; min-height:0;">
            <div id="plot6_" class="plotly-container" style="width:100%; height:100%;"></div>
        </div>
    </div>

    <style>
    .tabs-group {{
        background: #2D6A4F;
        padding: 12px;
        border-radius: 8px;
        height: 100%;
        display: flex;
        flex-direction: column;
    }}

    .tab-buttons {{
        display: grid;
        grid-template-columns: repeat(auto-fit, minmax(120px, 1fr));
        gap: 8px;
        margin-bottom: 12px;
        flex-shrink: 0;
    }}

    .tab-buttons button {{
        padding: 10px 12px;
        border: none;
        background: #1B4332;
        color: white;
        cursor: pointer;
        border-radius: 6px;
        transition: 0.3s;
        font-size: 14px;
        white-space: nowrap;
        overflow: hidden;
        text-overflow: ellipsis;
    }}

    .tab-buttons button:hover {{
        background: #666;
    }}

    .tab-content-secondary {{
        display: none;
        background: #f5f5f5;
        padding: 15px;
        border-radius: 8px;
        flex: 1;
        min-height: 0;  /* Importante para flexbox */
        width: 100%;
        box-sizing: border-box;
    }}

    .tab-content-secondary.active {{
        display: block;
    }}

    .plotly-container {{
        width: 100%;
        height: 100%;
        min-height: 300px;  /* Altura m√≠nima de respaldo */
    }}

    /* Asegurar que los divs de Plotly ocupen todo el espacio */
    .plotly-container .js-plotly-plot,
    .plotly-container .plotly,
    .plotly-container .plot-container {{
        width: 100% !important;
        height: 100% !important;
    }}
    </style>

    <script src="https://cdn.plot.ly/plotly-2.30.0.min.js"></script>    
    <script>
        // Configuraci√≥n de las figuras
        var fig1 = {plot_colombia1_json};
        var fig2 = {plot_colombia2_json};
        var fig3 = {plot_colombia3_json};
        var fig4 = {plot_colombia4_json};
        var fig5 = {plot_colombia5_json};
        var fig6 = {plot_colombia6_json};

        // Asegurar que las figuras tengan autosize=True y m√°rgenes adecuados
        [fig1, fig2, fig3, fig4, fig5, fig6].forEach(function(fig) {{
            if (!fig.layout) fig.layout = {{}};
            fig.layout.autosize = true;
            fig.layout.width = null;
            fig.layout.height = null;
            fig.layout.margin = {{l: 40, r: 40, t: 40, b: 40}};
        }});

        // Mapa de figuras por tab
        var figs = {{
            tab1_: {{data: fig1.data, layout: fig1.layout, div: "plot1_"}},
            tab2_: {{data: fig2.data, layout: fig2.layout, div: "plot2_"}},
            tab3_: {{data: fig3.data, layout: fig3.layout, div: "plot3_"}},
            tab4_: {{data: fig4.data, layout: fig4.layout, div: "plot4_"}},
            tab5_: {{data: fig5.data, layout: fig5.layout, div: "plot5_"}},
            tab6_: {{data: fig6.data, layout: fig6.layout, div: "plot6_"}},
        }};

        // Control de gr√°ficos ya plotteados
        var plotted = {{
            tab1_: false,
            tab2_: false,
            tab3_: false,
            tab4_: false,
            tab5_: false,
            tab6_: false
        }};

        // Funci√≥n para forzar el resize de un plot espec√≠fico
        function forcePlotResize(divId) {{
            var plotDiv = document.getElementById(divId);
            if (plotDiv && plotDiv.layout) {{  // Verificar si es un plot de Plotly
                Plotly.Plots.resize(plotDiv);
            }}
        }}

        function openTabGroup(groupId, tabId) {{
            // Obtener el grupo
            const group = document.getElementById(groupId);
            if (!group) return;

            // Cerrar SOLO las pesta√±as de ese grupo
            group.querySelectorAll(".tab-content-secondary")
                .forEach(tab => tab.classList.remove("active"));

            // Abrir la seleccionada
            const selectedTab = document.getElementById(tabId);
            if (selectedTab) {{
                selectedTab.classList.add("active");
            }}

            // Plotear el gr√°fico si no se ha hecho antes
            setTimeout(function() {{
                if (figs[tabId] && !plotted[tabId]) {{
                    Plotly.newPlot(
                        figs[tabId].div,
                        figs[tabId].data,
                        figs[tabId].layout,
                        {{responsive: true}}
                    ).then(function() {{
                        plotted[tabId] = true;
                        // Forzar resize despu√©s de plotear
                        setTimeout(function() {{
                            forcePlotResize(figs[tabId].div);
                        }}, 100);
                    }});
                }} else if (figs[tabId] && plotted[tabId]) {{
                    // Si ya est√° plotteado, forzar resize
                    forcePlotResize(figs[tabId].div);
                }}
                
                // Tambi√©n disparar el resize global
                setTimeout(function() {{
                    window.dispatchEvent(new Event('resize'));
                }}, 200);
                
            }}, 100);
        }}

        // Inicializar primera pesta√±a cuando cargue la p√°gina
        window.addEventListener("DOMContentLoaded", function() {{
            setTimeout(function() {{
                openTabGroup('groupB', 'tab1_');
            }}, 200);
        }});

        function resizePlot(plot){{
            if (!plot || !plot.layout) return;  // Verificar que sea un plot de Plotly
            
            const parent = plot.parentElement;
            if (!parent) return;
            
            const rect = parent.getBoundingClientRect();
            
            // Solo hacer resize si el plot est√° visible
            if (rect.width > 0 && rect.height > 0) {{
                // 1) Forzar altura real en el div plotly
                plot.style.height = rect.height + "px";
                plot.style.width  = rect.width + "px";
                
                // 2) Ahora s√≠ Plotly recalcula correctamente
                try {{
                    Plotly.Plots.resize(plot);
                }} catch(e) {{
                    console.log("Error resizing plot:", e);
                }}
            }}
        }}

    function viewx_resize_all(){{
        document.querySelectorAll(".plotly-graph-div").forEach(function(plot) {{
            // Solo redimensionar plots visibles
            if (plot && plot.closest('.tab-content-main.active')) {{
                resizePlot(plot);
            }}
            if (plot && plot.closest('.tab-content-secondary.active')) {{
                resizePlot(plot);
            }}
        }});
    }}


    function attach(){{
        const ro = new ResizeObserver(entries => {{
            for (const entry of entries) {{
                // Solo procesar si el elemento est√° visible
                if (entry.target.offsetParent !== null) {{
                    const plots = entry.target.querySelectorAll(".plotly-graph-div");
                    plots.forEach(function(plot) {{
                        if (plot.closest('.tab-content-secondary.active')) {{
                            resizePlot(plot);
                        }}
                        if (plot.closest('.tab-content-main.active')) {{
                            resizePlot(plot);
                        }}
                    }});
                }}
            }}
        }});

        document.querySelectorAll(".viewx-slot").forEach(el => ro.observe(el));

        // primera ejecuci√≥n (MUY importante)
        setTimeout(viewx_resize_all, 200);
        
        // Tambi√©n observar cambios en las pesta√±as
        const observer = new MutationObserver(function(mutations) {{
            mutations.forEach(function(mutation) {{
                if (mutation.target.classList && 
                    mutation.target.classList.contains('tab-content-main')) {{
                    setTimeout(viewx_resize_all, 100);
                }}
                if (mutation.target.classList && 
                    mutation.target.classList.contains('tab-content-secondary')) {{
                    setTimeout(viewx_resize_all, 100);
                }}
            }});
        }});
        
        document.querySelectorAll('.tabs-group').forEach(function(group) {{
            observer.observe(group, {{ 
                attributes: true, 
                attributeFilter: ['class'],
                subtree: true 
            }});
        }});
    }}

    window.addEventListener("DOMContentLoaded", attach);
    window.addEventListener("resize", function() {{
        setTimeout(viewx_resize_all, 100);
    }});
    </script>
    
    </div>
    """,
    slot_grid=("div5", 2, 7, 2, 3)
)
```
